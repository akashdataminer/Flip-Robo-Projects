# Evaluating Model using different Regressors:-

### Linear Regression:-

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Importing the required libraries and functions from sklearn
```

```python
MaxAccu=0
MaxRS=0

for i in range (1,500):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    predlr_train=lr.predict(x_train)
    predlr_test=lr.predict(x_test)
    acc=r2_score(y_test,predlr_test)
    if acc>MaxAccu:
        MaxAccu=acc
        MaxRS=i
print("Best Accuracy is ", MaxAccu, "on Random State ", MaxRS)

# Finding the Best Random State with Maximum Accuracy
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=181)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)

# Training the linear regression model
```

```python
lr_pred=lr.predict(x_test)
lr_pred

# The Linear Regression model is making predictions based on training
```

```python
print(r2_score(y_test, lr_pred))

# Printing the Final Accuracy Score of Linear Regression Model
```

### Random Forest Regressor:-

```python
from sklearn.ensemble import RandomForestRegressor

# Importing the Random Forest Regressor Model from sklearn Library
```

```python
MaxAccu=0
MaxRS=0

for i in range (1,500):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=i)
    rfr=RandomForestRegressor()
    rfr.fit(x_train,y_train)
    predrfr_train=rfr.predict(x_train)
    predrfr_test=rfr.predict(x_test)
    acc=r2_score(y_test,predrfr_test)
    if acc>MaxAccu:
        MaxAccu=acc
        MaxRS=i
print("Best Accuracy is ", MaxAccu, "on Random State ", MaxRS)

# Finding the Best Random State with Maximum Accuracy
```
