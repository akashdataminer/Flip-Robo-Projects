```python
from sklearn.preprocessing import LabelEncoder

# Importing the Label Encoder from sklearn Library
```

```python
le=LabelEncoder()
```

```python
for i in cust.columns:
    if cust[i].dtypes=="object":
        cust[i]=le.fit_transform(cust[i].values.reshape(-1,1))
        
# Running a for loop to give numeric values to strings using the Label Encoder in training data
```
